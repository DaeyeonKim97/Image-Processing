```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
import math
```

```python
# 이미지를 확대하세요.
def my_resizing(img):
    matrix = np.array([[ 1.5 , 0 , 0 ],[ 0  , 1.5 , 0 ]], dtype=np.float32)                   # 1.5배 확대하는 matrix 만들기
    result = cv2.warpAffine(img, matrix, None)
    return result

resized = my_resizing(I)
cv2_imshow(resized)
```

```python
# 이미지를 가로: 30px, 세로: 20px 이동하세요
def my_translation(img):
    matrix = np.array([[ 1 , 0 , 30 ],[ 0 , 1 , 20 ]], dtype=np.float32)                # 가로와 세로를 이동하는 matrix 만들기
    result = cv2.warpAffine(img, matrix, None)
    return result

translation = my_translation(resized)
cv2_imshow(translation)
```

```python
# 이미지를 30도 회전하세요 (회전축은 이미지의 중심) - 실습과제 PPT를 참고하세요
def my_rotation(img, angle):
    center_x = img.shape[1]/2
    center_y = img.shape[0]/2
    
    rad = math.radians(angle)
    a = math.cos(rad)
    b = math.sin(rad)
    matrix = np.array([[ a , b , (1-a)*center_x -b*center_y  ],
                       [ -b , a , b*center_x + (1-a)*center_y  ]], dtype=np.float32)
    result = cv2.warpAffine(img, matrix, None)
    return result

rotation = my_rotation(translation, 30)
cv2_imshow(rotation)
```

```python
# 이미지를 shearing 하세요
def my_shear(img):
    
    matrix_x = np.array([[ 1 , 0.2 , 0 ],[ 0 , 1 , 0 ]], dtype=np.float32)                     # 가로: 0.2의 비율로 shearging
    matrix_y = np.array([[ 1 , 0 , 0 ],[ 0.2 , 1 , 0 ]], dtype=np.float32)                     # 세로: 0.2의 비율로 shearging
    
    result_1 = cv2.warpAffine(img, matrix_x, None)
    result_2 = cv2.warpAffine(result_1, matrix_y, None)
    return result_2

shearing = my_shear(rotation)
cv2_imshow(shearing)
```

```python
# 실험 결과를 아래 Plotting 하세요

fig = plt.figure(figsize = (15, 10)) # rows*cols 행렬의 i번째 subplot 생성
rows = 2
cols = 3

## plotting 하는 부분은 1주차 과제를 참조하여 스스로 채워 넣으세요. 

ax1 = fig.add_subplot(rows, cols, 1)
I_RGB = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)
ax1.imshow(I_RGB)
ax1.set_title('Original')
ax1.axis("off")

ax2 = fig.add_subplot(rows, cols, 2)
resized_RGB = cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)
ax2.imshow(resized_RGB)
ax2.set_title('Image Enlaged')
ax2.axis("off")

ax3 = fig.add_subplot(rows, cols, 3)
translation_RGB = cv2.cvtColor(translation, cv2.COLOR_BGR2RGB)
ax3.imshow(translation_RGB)
ax3.set_title('(Image Translated')
ax3.axis("off")

ax4 = fig.add_subplot(rows, cols, 4)
rotation_RGB = cv2.cvtColor(rotation, cv2.COLOR_BGR2RGB)
ax4.imshow(rotation_RGB)
ax4.set_title('Image Rotated')
ax4.axis("off")

ax5 = fig.add_subplot(rows, cols, 5)
shearing_RGB = cv2.cvtColor(shearing, cv2.COLOR_BGR2RGB)
ax5.imshow(shearing_RGB)
ax5.set_title('Image Shearing')
ax5.axis("off")

plt.show()
```
